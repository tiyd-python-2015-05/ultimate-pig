```python
from two_player import Game, Player, Challenger, Percentages
import matplotlib.pyplot as plt
import math
import statistics as st
```

```python
%matplotlib inline
```

```python
def simulation(n, player1_type, player2_type=Player):
    player1 = player1_type()
    player2 = player2_type()
    game = Game(player1, player2)
    game.run(games=n)
    return game.record[0]-game.record[1]
```

```python
def n_trials(player1_type, player2_type=Player, trials=10000, n=100):
    differences = []
    player1 = player1_type()
    player2 = player2_type()
    for trial in range(trials):
        game = Game(player1, player2)
        game.run(games=n)
        differences.append(game.record[0]-game.record[1])
    return differences
```

```python
default_differences = n_trials(Player)
Challenger_differences = n_trials(Challenger)
Percentages_differences = n_trials(Percentages)
```

```python
mean = st.mean(default_differences)
stdev = st.pstdev(default_differences)

plt.hist(default_differences, bins=41)
plt.ylabel("Frequency")
plt.xlabel("Mean Difference")
plt.title("Differences for Default Player vs Default Player across 10,000 Trials of 100 Games Each")
ymin, ymax = plt.ylim()
ymax = ymax + 100
plt.ylim(ymin, ymax)
plt.vlines(mean, ymin, ymax, color="red")
plt.vlines([mean - 2 * stdev, mean - stdev, mean + stdev, mean + 2 * stdev], ymin, ymax, linestyles="dashed", color="red")
plt.text(30, 600, r'Mean: ${}$'.format(str(round(mean, 4))), fontsize=10)
plt.text(30, 500, r'Std Dev: ${}$'.format(str(round(stdev, 4))), fontsize=10)
plt.subplots_adjust(left=2, right=3.5)
```

```python
mean2 = st.mean(Challenger_differences)
stdev2 = st.pstdev(Challenger_differences)

plt.hist(Challenger_differences, bins=37)
plt.ylabel("Frequency")
plt.xlabel("Mean Difference")
plt.title("Differences for Challenger Player vs Default Player across 10,000 Trials of 100 Games Each")
ymin, ymax = plt.ylim()
ymax = ymax + 100
plt.ylim(ymin, ymax)
plt.vlines(mean2, ymin, ymax, color="red")
plt.vlines([mean2 - 2 * stdev2, mean2 - stdev2, mean2 + stdev2, mean2 + 2 * stdev2], ymin, ymax, linestyles="dashed", color="red")
plt.text(25, 600, r'Mean: ${}$'.format(str(round(mean2, 4))), fontsize=10)
plt.text(25, 500, r'Std Dev: ${}$'.format(str(round(stdev2, 4))), fontsize=10)
plt.subplots_adjust(left=2, right=3.5)
```
