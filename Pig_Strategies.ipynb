###Appears that trying for 15 or 20 produce good results, as does 4 to 6 roll strategy.
Maybe combining the logic would optimize.
Roll up to 4 rolls, stop when you get to 20
or, roll up to 3 rolls, stop when you get to 15

```python
class Four_and_15_Rolls_Player(Player):
    def roll_again_logic(self):
        if self.roll_count == 4:
            return False
        elif self.tally < 15:
            return True
        else:
            return False
```

```python
trial_4_roll_15_player = n_trials(1000, Four_and_15_Rolls_Player)
```

```python
print((trial_4_roll_15_player)[:4])
```

###Conclusion
Conclusion:  Rolling to 20 or 25 or Rolling 4 to 6 times seem to be the best strategy.  Combining the two strategies by stopping at 22 if you get there before you roll 5 times or stopping at 27 if you get there before 6 rolls might work well. 6 to get 27 seems to have a nice amount of high outliers and a quartile range that is a little higher wider.  There is less upside for the rolls that always stop at 4.

You may want to decide if you want to go for a high score or go for a consistent score.
That will help you decide which is the best score.  For competition against another person constistency might be better.  I might recommend 4 rolls or 5 rolls, stopping at 22.
