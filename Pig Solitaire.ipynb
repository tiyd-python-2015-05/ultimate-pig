#Ultimate Pig Solitaire
Monte Carlo simulation of a game of pig solitaire that tests different strategies for playing.

```python
from ultimate_pig import Game, Player, Rolls5, Until18, PushIt, Rolls4
import matplotlib.pyplot as plt
import statistics as st
```

```python
%matplotlib inline
```

```python
def simulation(n, player_type):
    player = player_type()
    game = Game(player)
    scores = game.run(games=n)
    return sum(scores)/len(scores)
```

```python
def n_trials(player=Player, trials=10000, games=100):
    mean_scores = []
    for trial in range(trials):
        mean_scores.append(simulation(games, player))
    return mean_scores
        
```

```python
default_means = n_trials()
rolls5_means = n_trials(Rolls5)
until18_means = n_trials(Until18)
push_it_means = n_trials(PushIt)
rolls4_means = n_trials(Rolls4)
```

```python
class Until20(Player):
    def decide(self, score):
        if score < 20:
            return True
        else:
            return False
```

```python
until20_means = n_trials(Until20)
```

```python
mean1 = st.mean(default_means)
stdev1 = st.pstdev(default_means)
plt.subplot(6,1,1)
plt.hist(default_means, bins=40)
plt.ylabel("Frequency")
plt.xlabel("Mean Score")
plt.title("Mean Scores for Default Player across 10,000 Trials of 100 Games Each")
ymin, ymax = plt.ylim()
ymax += 50
plt.ylim(ymin, ymax)
plt.vlines(mean1, ymin, ymax, color="red")
plt.vlines([mean1 - 2 * stdev1, mean1 - stdev1, mean1 + stdev1, mean1 + 2 * stdev1], ymin, ymax, linestyles="dashed", color="red")
plt.text(24.5, 600, r'Mean: ${}$'.format(str(round(mean1, 4))), fontsize=10)
plt.text(24.5, 500, r'Std Dev: ${}$'.format(str(round(stdev1, 4))), fontsize=10)

mean2 = st.mean(rolls5_means)
stdev2 = st.pstdev(rolls5_means)
plt.subplot(6,1,2)
plt.hist(rolls5_means, bins=40)
plt.ylabel("Frequency")
plt.xlabel("Mean Score")
plt.title("Mean Scores for Rolls5 Player across 10,000 Trials of 100 Games Each")
ymin, ymax = plt.ylim()
plt.vlines(mean2, ymin, ymax, color="red")
plt.vlines([mean2 - 2 * stdev2, mean2 - stdev2, mean2 + stdev2, mean2 + 2 * stdev2], ymin, ymax, linestyles="dashed", color="red")
plt.text(63, 600, r'Mean: ${}$'.format(str(round(mean2, 4))), fontsize=10)
plt.text(63, 500, r'Std Dev: ${}$'.format(str(round(stdev2, 4))), fontsize=10)

mean3 = st.mean(until18_means)
stdev3 = st.pstdev(until18_means)
plt.subplot(6,1,3)
plt.hist(until18_means, bins=40)
plt.ylabel("Frequency")
plt.xlabel("Mean Score")
plt.title("Mean Scores for Until18 Player across 10,000 Trials of 100 Games Each")
ymin, ymax = plt.ylim()
plt.vlines(mean3, ymin, ymax, color="red")
plt.vlines([mean3 - 2 * stdev3, mean3 - stdev3, mean3 + stdev3, mean3 + 2 * stdev3], ymin, ymax, linestyles="dashed", color="red")
plt.text(63, 600, r'Mean: ${}$'.format(str(round(mean3, 4))), fontsize=10)
plt.text(63, 500, r'Std Dev: ${}$'.format(str(round(stdev3, 4))), fontsize=10)

mean4 = st.mean(push_it_means)
stdev4 = st.pstdev(push_it_means)
plt.subplot(6,1,4)
plt.hist(push_it_means, bins=40)
plt.ylabel("Frequency")
plt.xlabel("Mean Score")
plt.title("Mean Scores for PushIt Player across 10,000 Trials of 100 Games Each")
ymin, ymax = plt.ylim()
plt.vlines(mean4, ymin, ymax, color="red")
plt.vlines([mean4 - 2 * stdev4, mean4 - stdev4, mean4 + stdev4, mean4 + 2 * stdev4], ymin, ymax, linestyles="dashed", color="red")
plt.text(63, 600, r'Mean: ${}$'.format(str(round(mean4, 4))), fontsize=10)
plt.text(63, 500, r'Std Dev: ${}$'.format(str(round(stdev4, 4))), fontsize=10)
plt.subplots_adjust(left=2, right=3.5, top=5.5)

mean5 = st.mean(rolls4_means)
stdev5 = st.pstdev(rolls4_means)
plt.subplot(6,1,5)
plt.hist(rolls4_means, bins=40)
plt.ylabel("Frequency")
plt.xlabel("Mean Score")
plt.title("Mean Scores for Rolls4 Player across 10,000 Trials of 100 Games Each")
ymin, ymax = plt.ylim()
plt.vlines(mean5, ymin, ymax, color="red")
plt.vlines([mean5 - 2 * stdev5, mean5 - stdev5, mean5 + stdev5, mean5 + 2 * stdev5], ymin, ymax, linestyles="dashed", color="red")
plt.text(61, 600, r'Mean: ${}$'.format(str(round(mean5, 4))), fontsize=10)
plt.text(61, 500, r'Std Dev: ${}$'.format(str(round(stdev5, 4))), fontsize=10)

mean6 = st.mean(until20_means)
stdev6 = st.pstdev(until20_means)
plt.subplot(6,1,6)
plt.hist(until20_means, bins=40)
plt.ylabel("Frequency")
plt.xlabel("Mean Score")
plt.title("Mean Scores for Until20 Player across 10,000 Trials of 100 Games Each")
ymin, ymax = plt.ylim()
plt.vlines(mean6, ymin, ymax, color="red")
plt.vlines([mean6 - 2 * stdev6, mean6 - stdev6, mean6 + stdev6, mean6 + 2 * stdev6], ymin, ymax, linestyles="dashed", color="red")
plt.text(65, 600, r'Mean: ${}$'.format(str(round(mean6, 4))), fontsize=10)
plt.text(65, 500, r'Std Dev: ${}$'.format(str(round(stdev6, 4))), fontsize=10)
plt.show()
```

```python
plt.boxplot([default_means, rolls5_means, until18_means, push_it_means, rolls4_means,
            until20_means])
plt.title("Pig Strategies")
plt.ylabel("Mean Scores")
plt.xticks(range(1,7), ["Rolls Once", "Rolls Five Times", "Rolls Until a Score of 18", 
                       "Rolls 10 Times or Until a Score of 30", "Rolls 4 Times", 
                        "Rolls Until a Score of 20"])
plt.subplots_adjust(left=2, right=4.5, top=3.5, bottom=1.5)
plt.show()
```

#Conclusions
It would appear that as far as strategies for pig solitaire are concerned, the margin between various strategies is often very slim.  Of the strategies tested here, rolling until one reaches a score of 20 or more and then stopping appears to be the best option with a mean score of 56.9 and standard deviation of 2.8.  That slightly edges out the mean score of 56.5 for rolling until you reach 18, although rolling to 18 has a smaller standard deviation.  Both means lie within the standard deviation of the other, so it is hard to claim that one is diffinitively better than the other, but that small margin might just make the difference between breaking your high score or not.
