```python
from strategies import simple_threshold, complex_threshold
from pig import Pig
import random
from players import Player, SmartPlayer
from matplotlib import pyplot as plt
%matplotlib inline
import seaborn
import statistics as st
```

```python
pig = Pig(score_lim=100)
player1 = Player("Random")
player2 = SmartPlayer("Gilgamesh", simple_threshold(.5))
pig.set_players(player1, player2)
```

```python
def run_game(pig, num):
    for _ in range(num):
        pig.loop()
        pig.new_game()
```

```python
run_game(pig, 100)
```

```python
def winnings(pig):
    return [player.wins for player in pig.players]
    
        
print(winnings(pig)) # Gilgamesh wins almost every time against random
```

```python
pig.clear_players()
players = ["simple", "simple"]
thresholds = [.5, random.random()]
pig.create_players(players, thresholds)
```

```python
def unzip(items):
    return [[item[idx] for item in items] for idx in range(len(items[0]))]
        
```

```python
final = {}
for _ in range(10000):
    thresholds[1] = random.random() # different threshold each time
    pig.clear_players()
    pig.create_players(players, thresholds)
    run_game(pig, 100)
    final[thresholds[1]] = winnings(pig)
```

```python
entries = unzip([(item, final[item][1]/100) for item in final if final[item][1] != 100])
plt.scatter(*entries)
plt.text(.6, .8, "max: {}".format(round(max(entries[1]), 3)))
plt.text(.6, .7, "at x = {}".format(round(entries[0][entries[1].index(max(entries[1]))], 2)))
```
