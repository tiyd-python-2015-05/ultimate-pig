```python
def pig_trials(player1, player2, num_trials=1000):
    winners = []
    for _ in range(num_trials):
        game = PigGame(player1, player2)
        winners.append(game.play())
#     print('num_trials: {}  Mean: {}  Stdev: {}'.format(
#             num_trials,st.mean(scores), st.stdev(scores)))
    player1_wins = [player for player, score in winners].count(player1)
    player2_wins = num_trials - player1_wins
    player1_win_loss_ratio = round(player1_wins/player2_wins, 3)
    player2_win_loss_ratio = round(player2_wins/player1_wins, 3)
    print('Number of Trials: {} Player 1 win/loss ratio: {} Player 2 win/loss ratio: {}'.format(
        num_trials, player1_win_loss_ratio, player2_win_loss_ratio))
    return winners
```

```python
def plot_y_hist(data, title = "Plot"):
    plt.hist(data, bins=20)
    plt.xlabel("score")
    plt.ylabel("frequency")
    plt.title(title)

    ymin, ymax = plt.ylim()
    mean = st.mean(data)
    stdev = st.stdev(data)
    plt.vlines(mean, ymin, ymax, color="red")
    plt.vlines([mean - 2 * stdev, mean - stdev, mean + stdev, mean + 2 * stdev], ymin, ymax, linestyles="dashed", color="red")
    plt.show()
```

##Conclusions
###Probability Analysis
On any given roll, there is a 1/6 probability of rolling a 1, and thus a 5/6 probability of rolling anything else. After 2 rolls, the probability of not yet rolling a 1 is thus (5/6)^2, and (5/6)^3 for 3 rolls and so on. The list comprehension below shows this probabiity for 1 to 10 turns, with the probability falling below 0.5 between 3 and 4 rolls.

```python
non_1_probability = [(5/6)**n for n in range(11)]
print([round(n,3) for n in non_1_probability][1:])
```

###Simulation Results
In running 100,000 trials using basic strategies of holding after n number of rolls, several trends were observed. The standard deviation increased with the number of rolls, and the histograms showed a broadening as the number of rolls increased. The mean ending score also increased with increasing number of rolls until approximately 5-6 rolls, where the increases started to level off. Although the mean increased with increasing number of rolls, the histograms also become more and more skewed toward the low numbers, with the bar representing ~0 final score increasing steadily with number of rolls
